## Trasnformação de Dados

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 1 com astype

df_contracts.TotalCharges = df_contracts.TotalCharges.astype(float)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 2 com to_numeric
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 3 com to_numeric com coerce (trasforma em null quando não for valido a transformação)
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges, errors = 'coerce')
```

```python
# Mostrar estrutura / schema DataFrame
df_contracts.info()
```

Após a transformação da coluna TotalCharges do Dataframe Contracts , a coluna passou a ter valores ausentes (missing values)

## Unificando DataFrames de Customers, Services e Contracts


```python
# Contar quantidade de registros por DataFrame 
len(df_customers)
```

```python
len(df_services)
```

```python

len(df_contracts)
```

```python
# Renomear colunas em comun dos Dataframes para realizar os merges 
df_services.rename(columns={"customerID": "IDCliente"}, inplace=True)
```

```python
df_services.info()
```

```python
# Unificar DataFrame costumers com services, criando um terceiro DataFrame

df_temp = df_customers.merge(df_services, on =['IDCliente'])
```

```python
#Mostrar estrutura

df_temp.info()
```

```python
# Unificar o df_temp com Contracts, usando colunas de junçao com nomes distintos

df_churn_temp = df_temp.merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Unificar os tres DataFrame ao mesmo tempo com colunas com nomes diferentes

df_churn = df_customers.merge(df_services, on = ['IDCliente']).merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
#Removendo coluna de um DataFrame
df_churn.drop(['customerID'], axis = 1 , inplace = True)
```

```python
df_churn.info()
```
