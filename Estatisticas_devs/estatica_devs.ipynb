## CRIAÇÃO DATA FRAME

```python
import pandas as pd 
```

```python
pd.__version__
```

```python
dic_medidas = {'idade': [15, 18, 25, 25, 40, 55, 58, 60, 80], 'altura': [160, 162 , 165, 168, 172, 174 , 174, 174 ,176 ]}
```

```python
dic_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dic_medidas)
```

```python
df_medidas
```

## MEDIDA RESUMO


### MEDIDAS DE POSIÇÃO

```python
#meida
df_medidas['idade'].mean()
```

```python
#mediana
df_medidas['idade'].median()
```

```python
#mode 
df_medidas['idade'].mode()
```

### MEDIDAS DE DISPERÇÃO

```python
# variâncias 
df_medidas.idade.var()
```

```python
#desvio padrão
df_medidas.idade.std()
```

```python
# coeficiente de variação

df_medidas.idade.std() / df_medidas.idade.mean() * 100 
```

```python
df_medidas.altura.std() / df_medidas.altura.mean() * 100 
```

##### A variaçao da idade é muito maior doque a altura de acordo com o coeficiente de variação, uma sendo 3,4% e outro 54%

### MEDIDAS DE FORMAS 


```python
#Assimetria - Idade
df_medidas.idade.skew()
```

```python
#Assimetria - Altura
df_medidas.altura.skew()
```

```python
# Curtose - Idade
df_medidas.idade.kurtosis()
```

```python
# Curtose - Altura
df_medidas.altura.kurtosis()
```

```python
# Obter medidas estaticas 
df_medidas.idade.describe()
```

## CORELAÇÃO

```python
#Corelaçao do DataFrame, segundo pearson
df_medidas.corr(method="pearson")
```

```python
# Corelaçao 2 variaveis

df_medidas.idade.corr(df_medidas.altura)
```

```python
#Corelaçao do DataFrame, segundo pearson
df_medidas.corr(method="spearman")
```

## Represntações Gráficas

### Criar Data Frame de Vendas

```python
dic_vendas = {'categoria': ['feminino', 'masculino', 'infantil','casa'], 'valor':[400000, 600000, 250000, 580000], 'quantidade': [3000, 5000,1500, 2500]}
```

```python
dic_vendas
```

```python
df_vendas = pd.DataFrame.from_dict(dic_vendas)
```

```python
df_vendas
```

```python
#Grafico de barras vertical - Valor
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
#Grafico de barras vertical - Valor (ordenado)
df_vendas.sort_values('valor', ascending=False ).plot.bar(x='categoria',y='valor')
```

```python
#Grafico de barras horizontal - Quantidade

df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
#Grafico de barras horizontal - Quantidade (Ordenado - Ranking)
df_vendas.sort_values('quantidade').plot.barh(x='categoria', y='quantidade')
```

```python
# Grafico de disperção

df_medidas.plot.scatter(x='idade',y= 'altura')
```

```python
# Grafico Boxplot - Idade

df_medidas.idade.plot.box()
```

```python
# Grafico de Boxplot - Altura

df_medidas.altura.plot.box()
```

```python
dict_faturamento
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
df_faturamento.info()
```

```python
#Grafico de Linhas
df_faturamento.plot.line(x='data_ref',y='valor')
```
